```python
n_rows = 10
n_cols = 5 
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                 columns = cols) 
df
```

```python
df[(df['A'] == 1) | (df["A"] == 3)]
#operator or "|" = fungsi untuk menampilkan baris yang di inginkan
```

```python
df[df['A'].isin([1,3])]
#fungsi isin() = versi lebih simple dari fungsi or
```

```python
df[~df["A"].isin([1,3])]
#negasi "~" = fungsi untuk menampilkan baris yang nilai nya tidak sama dengan nilai yang di inginkan
```

```python
data = {'cols1':['1', '2', '3', 'teks'], 'cols2':['1', '2', '3', '4']}

df = pd.DataFrame(data)
df
```

```python
df.apply(pd.to_numeric, errors='coerce')
#to_numeric() = mengubah type data yang di inginkan menjadi numeric,
#untuk type data yang bukan numeric menggunakan "coerce" type data akan menjadi NaN(Non a Number)
```

```python
n_rows = 5
n_cols = 2
cols = ['bil pecahan', 'bil bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df['bil pecahan'] = df['bil pecahan'].astype('float')
#untuk bilangan pecahan di konversi menajdi float menggunakan fungsi .astype

df.index = pd.util.testing.makeDateIndex(n_rows, freq='min')
df = df.reset_index()

df['teks'] = list('ABCDE')
df
```

```python
df.select_dtypes(include='int')
#.select_dtypes(include='...') = berfungsi untuk menunjukan type data yang di inginkan
```
